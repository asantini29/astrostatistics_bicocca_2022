```python
import numpy as np
import matplotlib.pyplot as plt

rc = {"font.family" : "serif", 
      "mathtext.fontset" : "cm",
      "font.size" : "12",
      "figure.figsize" : [8,8]}
plt.rcParams.update(rc)

%config InlineBackend.figure_format='retina'
```

```python
N = int(1e5) #number of samples
low = 0.1 #lower value of the distribution
high = 10 #higher value of the distribution
x = np.random.uniform(low, high, N)
```

```python
plt.hist(x, density = True,histtype='step',lw=2)
prob_x = 1 / (high - low) #probability for an uniform distribution
plt.axhline(prob_x, c= 'r', label  = '$p(x)$')

plt.xlabel('$x$')
plt.ylabel('probability')
plt.legend(loc='lower center')
plt.show()
```

```python
y = np.log10(x)
plt.hist(y, density = True, histtype='step',lw=2)
plt.xlabel('$y$')
plt.ylabel('probability')
plt.show()
```

The derivative is:  
$\frac{dy}{dx} = \frac{1}{x \ln(10)} \implies \frac{dx}{dy} = x \ln(10) = 10^y \ln(10)$

```python
y_array = np.linspace(-1, 1, N)
plt.hist(y, density = True, histtype='step',lw=2)
prob_y = prob_x * np.log(10) * 10**y_array
plt.plot(y_array, prob_y, label = '$p(y)$')

plt.xlabel('$y$')
plt.ylabel('probability')
plt.legend()
plt.show()
```

```python
print('The log of the mean of x is: ', np.log10(np.mean(x)))
print('The mean of the log of x is: ', np.mean(y))
```

```python
print('The log of the median of x is: ', np.log10(np.median(x)))
print('The median of the log of x is: ', np.median(y))
```
