```python
import numpy as np
import pylab as plt
from scipy.stats import poisson
from tqdm.notebook import tqdm

rc = {"font.family" : "serif", 
      "mathtext.fontset" : "cm",
      "font.size" : "12",
      "figure.figsize" : [8,8]}
plt.rcParams.update(rc)

%config InlineBackend.figure_format='retina'
```

```python
deaths = [0, 1, 2, 3, 4]
groups = [109, 65, 22, 3, 1]
```

```python
total = np.sum(groups) #entire army
freq = groups / total #frequencies 
```

```python
plt.scatter(deaths, freq, s = 21)
plt.xlabel('deaths')
plt.ylabel('frequency')
plt.title('Frequencies')
plt.show()
```

looks like an exponential decrease $\implies$ **Poisson's distribution**

```python
#Poisson distribution: 
mean = np.average(deaths, weights = freq) #weighted average of the deaths
distr = poisson(mean)

plt.scatter(deaths, freq, s = 21, label = 'empirical frequencies')
plt.plot(deaths, distr.pmf(deaths), c='r', label = 'Poisson')

plt.xlabel('deaths')
plt.ylabel('frequency')
plt.title("Poisson's distribution")
plt.legend(frameon = False)
plt.show()
```
